```python
from pyspark.sql import SparkSession

# Crear sesi칩n Spark con el driver JDBC
spark = SparkSession.builder \
    .appName("PostgresConnectionTest") \
    .config("spark.jars", "/opt/spark/jars/postgresql-42.7.1.jar") \
    .getOrCreate()

# Leer tabla del sistema de PostgreSQL
df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://postgres:5432/sparkdb") \
    .option("dbtable", "pg_catalog.pg_tables") \
    .option("user", "sparkuser") \
    .option("password", "sparkpass") \
    .load()

df.show(5)
```
